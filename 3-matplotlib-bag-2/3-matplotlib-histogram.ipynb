Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Membuat df, dengan data 100 x 2, interval -100 sampai 100

```python
df = pd.DataFrame(np.random.randint(-100, 100, size=(100,2)), columns=list("AB"))
df
```

membuat histogram dengan tuple 'A'

```python
plt.figure()
df['A'].plot.hist()
```

membuat histogram dengan tuple 'A'

```python
df['A'].hist()
```

Membuat histogram dengan stacked atau bertumpuk

```python
df.plot.hist(stacked=True)
```

Menambah property alpha untuk melihat transparancy chart

```python
df.plot.hist(alpha=0.5)
```

Merubah chart menjadi horizontal dengan props orientation

```python
df['A'].plot.hist(orientation='horizontal')
```

membuat dict dengan index matkul dan jumlah mahasiswa yang mempunya list

```python
data = {
    'matkul' : ['matematika', 'biologi', 'fisika', 'kimia'],
    'jml_mhs' : [100, 120, 140, 90]}
df1 = pd.DataFrame(data)
df1
```

membuat pie chart menggunakan data diatas

```python
df1.plot(kind='pie', y = 'jml_mhs', labels=df1['matkul'], legend=True)
```
