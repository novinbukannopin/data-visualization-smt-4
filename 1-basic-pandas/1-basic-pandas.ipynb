Import Pandas Library

```python
import pandas as pd
```

Import Dataset kedalam DataFrame

```python
df = pd.read_csv("pokemon_data.csv")
df
```

Data Filtering

Get Nama Kolom di DataFrame

```python
print(df.columns)
```

Get Value dari Kolom Tertentu

```python
print(df['Name'])
```

Get Value dari Index x ke y
Misal [0:5] -> 0 - 4

```python
print(df['Name'][0:5])
```

Get Value dari banyak Kolom -> Array

```python
print(df[['Name', 'Type 1', 'HP']])
```

Get Info Semua Kolom, pada df berindex x

```python
print(df.iloc[0])
```

Get Data dengan nama x, berdasarkan kondisi y

```python
df.loc[df['Type 1'] == 'Grass']
```

Get Data dengan Multiple Condition -> using operator &

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

Get Value dari kolom x, contains y

```python
df.loc[df['Name'].str.contains('Mega')]
```

Sort kolom x, default = asc, desc => Ascending=False

```python
df.sort_values('Name')
```

Buat Kolom baru x, hasil operasi y + z + n

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df['Total']
```

Ubah value dari x ke y

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flame'
df
```

Export to csv, without index

```python
df.to_csv('output.csv', index=False)
```
