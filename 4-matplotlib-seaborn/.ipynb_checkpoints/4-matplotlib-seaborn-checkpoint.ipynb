Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Panggil dataset ke var df

```python
df = pd.read_csv('HousingData.csv')
```

Show head untuk df

```python
df.head()
```

Hitung sum kolom yang null

```python
df.isnull().sum()
```

Mofidy kolom dengan nama baru

```python
new_columns_dist = {
    'CRIM' : 'rate_of_crime',
    'ZN' : 'residential_zone_pct',
    'INDUS' : 'business_zone_pct',
    'CHAS' : 'borders_river',
    'NOX' : 'oxide_concentration',
    'RM' : 'average_rooms',
    'AGE' : 'owner_occup_pct',
    'DIS' : 'dist_to_work',
    'RAD' : 'access_to_highway',
    'TAX' : 'property_tax',
    'PTRATIO' : 'student_teacher_ratio',
    'LSTAT' : 'pct_underclass',
    'MEDV' : 'home_median_value'
}
df.rename(columns=new_columns_dist, inplace=True)
df.head()
```

Describe statistic dengan df

```python
df.describe()
```

Membuat boxplot dengan data rate of crime

```python
sns.boxplot(df['rate_of_crime'])
```

Membuat boxplot dengan data dist to work

```python
sns.boxplot(df['dist_to_work'])
```

Membuat figure 20 x 4
Membuat boxplot dengan x dan y itu

```python
plt.figure(figsize=(20,4))
sns.boxplot(x='variable', y='value', data=pd.melt(df))
```

Membuat heatmap, dengan data hasil korelasi

```python
plt.subplots(figsize=(20,4))
sns.heatmap(df.corr(), annot=True)
```

Menampilkan corelasi dari home_median_value secara asc

```python
df.corr()['home_median_value'].sort_values()
```

Membuat pairplot dari data df

```python
sns.pairplot(df)
```
